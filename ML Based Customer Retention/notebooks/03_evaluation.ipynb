```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

models = ['Random Forest', 'SVM', 'Naive Bayes', 'XGBoost', 'Logistic Reg']
y_preds = [rf.predict(X_test_scaled), y_pred_svm, y_pred_nb, y_pred_xgb, y_pred_log]

accuracy_scores = [accuracy_score(y_test, pred) for pred in y_preds]
precision_scores = [precision_score(y_test, pred, average='weighted') for pred in y_preds]
recall_scores = [recall_score(y_test, pred, average='weighted') for pred in y_preds]
f1_scores = [f1_score(y_test, pred, average='weighted') for pred in y_preds]

metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
x = np.arange(len(metrics))
width = 0.15

fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width*2, [accuracy_scores[0], precision_scores[0], recall_scores[0], f1_scores[0]], width, label='Random Forest')
rects2 = ax.bar(x - width, [accuracy_scores[1], precision_scores[1], recall_scores[1], f1_scores[1]], width, label='SVM')
rects3 = ax.bar(x, [accuracy_scores[2], precision_scores[2], recall_scores[2], f1_scores[2]], width, label='Naive Bayes')
rects4 = ax.bar(x + width, [accuracy_scores[3], precision_scores[3], recall_scores[3], f1_scores[3]], width, label='XGBoost')
rects5 = ax.bar(x + width*2, [accuracy_scores[4], precision_scores[4], recall_scores[4], f1_scores[4]], width, label='Logistic Reg')

ax.set_ylabel('Scores')
ax.set_title('Performance Metrics for Different Models')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)

plt.ylim(0, 1.1)
plt.tight_layout()
plt.show()
```
